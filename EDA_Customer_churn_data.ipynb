## Treatment of null values

1. Drop the null records
2. Fill the null records

* If the column is a discrete column then we fill the null values with mean or median.
* If the column is a categorical column then we fill the null values with mode.
* If any column has more than 30% values as null values then we drop that column.

### Outliers in numeric columns

```python
sns.boxplot(y = 'weekly_mins_watched', data=obj_num)
```

###Remove outliers from 'weekly mins watched'

```python
Q1=obj_num['weekly_mins_watched'].quantile(0.25)
Q3=obj_num['weekly_mins_watched'].quantile(0.75)
IQR=Q3-Q1
IQR
```

```python
#boundries

lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR

lower_limit, upper_limit
```

```python
#outlier list for 'weekly mins watched'

outlier_list_df = obj_num[((obj_num['weekly_mins_watched'] < lower_limit) | (obj_num['weekly_mins_watched'] > upper_limit))]
outlier_list_df
```

```python
#find median of 'weekly mins watched'

median_of_weekly_mins_watched = obj_num['weekly_mins_watched'].median()
median_of_weekly_mins_watched
```

```python
#replace outliers with median

obj_num.loc[(obj_num['weekly_mins_watched'] < lower_limit) | (obj_num['weekly_mins_watched'] > upper_limit), 'weekly_mins_watched'] = median_of_weekly_mins_watched
```

```python
outlier_list_df = obj_num[((obj_num['weekly_mins_watched'] < lower_limit) | (obj_num['weekly_mins_watched'] > upper_limit))]
outlier_list_df
```

```python
sns.boxplot(y = 'weekly_mins_watched', data=obj_num)
```

### Treatment of outliers for num_data

```python
cols = list(obj_num.columns)
cols

def out_data(obj_num):
    q1=obj_num.quantile(0.25)
    q3=obj_num.quantile(0.75)

    iqr=q3-q1


    lower=q1-1.5*iqr
    upper=q3+1.5*iqr

    return obj_num.clip(lower=lower,upper=upper)
```

```python
for i in cols:
    if i!='churn':
        obj_num[i]=out_data(obj_num[i])
```
